### Web Scraping Top Disposable Diaper Brand Names 

```python
URL = "https://www.amazon.com/gp/search/other/ref=lp_166772011_sa_p_89?rh=n%3A165796011%2Cn%3A%21165797011%2Cn%3A166764011%2Cn%3A166772011&bbn=166772011&pickerToList=lbr_brands_browse-bin&ie=UTF8&qid=1564048577"
driver = webdriver.Chrome('C:/Users/Stephane/Desktop/FTDS/Foundation/chromedriver_win32/chromedriver.exe')
driver.get(URL)
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
df_2 = pd.DataFrame(columns = ["Brand","Number of Products"])
while True:
    for i in soup.find_all(class_= "a-list-item"):
        try:
            Brand = i.find(class_= "refinementLink").text
        except:
            Brand = "Null"
        print(Brand)
        try:
            NumberOfProducts = i.find(class_= "narrowValue").text
        except:
            NumberOfProducts = "Null"
        print(NumberOfProducts)
        df_2 = df_2.append({"Brand":Brand, "Number of Products":NumberOfProducts}, ignore_index = True)
    else:
        break
```

```python
df = pd.read_csv("Amazon.csv")
```

```python
df2 = pd.read_csv("nappy_brands.csv")
```

```python
df2["Brand"] = df2["Brand"].str.upper()
BrandList = list(df2.Brand)
df["Product Name"] = df["Product Name"].str.upper()
```

```python
for x in range(0, len(df["Brand Name"])):
    for y in range(0, len(BrandList)):
        if (BrandList[y] in df["Product Name"].loc[x]) == True:
            df["Brand Name"].loc[x] = BrandList[y]
```

```python
df["Number of Reviews"].replace(',', '', regex = True, inplace = True)
df["Rating"].replace('Null', np.nan, regex = True, inplace = True)
df["Number of Reviews"].replace('Null', np.nan, regex = True, inplace = True)
df["Price"].replace('Null', np.nan, regex = True, inplace = True)
df["Price Count"].replace('Null', np.nan, regex = True, inplace = True)
df["Brand Name"].replace('', np.nan, regex = True, inplace = True)
```

```python
df["Rating"] = df["Rating"].astype(float)
df["Number of Reviews"] = df["Number of Reviews"].astype(float)
df["Price"] = df["Price"].astype(float)
df["Price Count"] = df["Price Count"].astype(float)
```
